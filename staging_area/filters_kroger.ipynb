```python
import pandas as pd
```

```python
# read csv
path='../kroger/kroger_milk_prices.csv'
df=pd.read_csv(path)
df
```

```python
filter_df=df.loc[df['categories'].str.contains('dairy', case=False)&df['description'].str.contains('milk', case=False)]
filter_df=filter_df.drop('Unnamed: 0', axis=1)
filter_df
```

```python
type_list=['almond','oat','chocolate','lactose','fat','whole','organic', 'powder','protein','other']
type_list
```

```python
milk_type_df=pd.DataFrame(columns=type_list)
milk_type_df['name']=filter_df['description']
milk_type_df['id']=filter_df['productId']
milk_type_df['types']=filter_df['description']
milk_type_df['type']=filter_df['description']
```

```python
#
for type in type_list:
    milk_type_df[type]=filter_df['description'].str.contains(type, case=False)
```

```python
milk_type_df
```

```python
for index in milk_type_df.index:
    milk_type_df["types"][index]=[]
    milk_type_df['other'][index]=True
    for type in type_list:
        if milk_type_df[type][index]:
            milk_type_df['other'][index]=False
            milk_type_df['types'][index].append(type)
    milk_type_df['type'][index]=milk_type_df['types'][index][0]
```

```python
milk_type_df
```

```python
milk_type_df=milk_type_df.drop(type_list, axis=1)
milk_type_df
```

```python
types_df=milk_type_df.groupby(['type']).agg({'id':'count'})
types_df=types_df.rename(columns={"id": "count_id"})
types_df
```

```python
types_df=types_df.reset_index(level=['type'])
types_df
```

```python
path='../data/kroger_milk_vs_types.csv'
milk_type_df.to_csv(path, index_label=False)
```

```python
path='../data/milk_types.csv'
types_df.to_csv(path, index_label=False)
```
