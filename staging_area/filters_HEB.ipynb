```python
import pandas as pd
```

```python
# read csv  for scrape process
path='../data/heb_scrape.csv'
heb_df=pd.read_csv(path)
heb_df
```

```python
# clean size of milk , removing extra characters
clean_heb_df=pd.DataFrame()
clean_heb_df=heb_df
clean_heb_df['size']=heb_df['size'].str.rsplit(pat=',').str[1]
clean_heb_df
```

```python
# # unit of measure sales price - per oz float 
# clean_heb_df['uomSalePrice']=heb_df['uomSalePrice'].str.strip('($/oz/fl )').astype(float)
# we will skip for now
```

```python
# convert features to list of features
clean_heb_df['features']=heb_df['features'].str.split(pat='.').str[:-1]
clean_heb_df
```

```python
# creating type column  and cleaning size and brand name out
clean_heb_df['type']=clean_heb_df['name'].str.rsplit(pat=',').str[0]
clean_heb_df['type']
```

```python
# in the loop remove brand name form milk name
for index in clean_heb_df.index:
    brand_length=len(clean_heb_df.loc[index, ('brand')])
    clean_heb_df.loc[index, ('type')]=clean_heb_df.loc[index, ('type')][brand_length:]
```

```python
# list of all names after cleaning 
clean_heb_df.groupby(['type']).agg({'type':'count'})
```

```python
# creating list of milk types for binning 
type_list=['almond','oat','chocolate','lactose','fat','whole','organic', 'powder','protein','other']
type_list
```

```python
# create milk_type DF to bin types
milk_type_df=pd.DataFrame(columns=type_list)
milk_type_df['name']=clean_heb_df['type']
milk_type_df['id']=clean_heb_df['id']
milk_type_df['types']=clean_heb_df['type']
milk_type_df['type']=clean_heb_df['type']
```

```python
# bining types - false /true matrix
for type in type_list:
    milk_type_df[type]=clean_heb_df['type'].str.contains(type, case=False)
```

```python
milk_type_df
```

```python
#combine matrix into list of types and pickup main type
for index in milk_type_df.index:
    milk_type_df['types'][index]=[]
    # making 'other' true , if no matches stays true
    milk_type_df.loc[index,('other')]=True
    for type in type_list:
        if milk_type_df.loc[index,(type)]:
            milk_type_df.loc[index,('other')]=False
            milk_type_df['types'][index].append(type)
    milk_type_df.loc[index,('type')]=milk_type_df.loc[index,('types')][0]
            
```

```python
milk_type_df
```

```python
# dropping matrix columns
heb_milk_type_df=milk_type_df.drop(type_list, axis=1)
heb_milk_type_df
```

```python
# creating milk types table 
types_df=heb_milk_type_df.groupby(['type']).agg({'id':'count'})
types_df=types_df.rename(columns={"id": "count_id"})
types_df
```

```python
heb_types_df=types_df.reset_index(level=['type'])
heb_types_df
```

```python
# output to csv 
path='../data/heb_clean.csv'
clean_heb_df.to_csv(path, index_label=False)
```
