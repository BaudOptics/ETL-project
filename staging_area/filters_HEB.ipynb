```python
import re
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# read csv
path='../data/heb_scrape.csv'
heb_df=pd.read_csv(path)
heb_df
```

```python
# clean size 
clean_heb_df=pd.DataFrame()
clean_heb_df=heb_df
clean_heb_df['size']=heb_df['size'].str.rsplit(pat=',').str[1]
clean_heb_df
```

```python
# convert features to list
clean_heb_df['features']=heb_df['features'].str.split(pat='.').str[:-1]
```

```python
clean_heb_df
```

```python
milk_names=clean_heb_df.groupby(['name']).agg({'id':'count'})
milk_names
```

```python
clean_heb_df['type']=clean_heb_df['name'].str.rsplit(pat=',').str[0]
clean_heb_df['type']
```

```python
for index in clean_heb_df.index:
    clean_heb_df['type'][index]=clean_heb_df['type'][index][len(clean_heb_df['brand'][index]):]
```

```python
clean_heb_df.groupby(['type']).agg({'type':'count'})
```

```python
clean_heb_df['simple_type']=clean_heb_df['type']
```

```python
type_list=['almond','oat','chocolate','lactose','fat','whole','organic', 'powder','protein','other']
type_list
```

```python
milk_type_df=pd.DataFrame(columns=type_list)
milk_type_df['name']=clean_heb_df['type']
milk_type_df['id']=clean_heb_df['id']
milk_type_df['types']=clean_heb_df['type']
milk_type_df['type']=clean_heb_df['type']
```

```python
milk_type_df['other'].all=True
for type in type_list:
    milk_type_df[type]=clean_heb_df['type'].str.contains(type, case=False)
```

```python
for index in milk_type_df.index:
    milk_type_df["types"][index]=[]
    milk_type_df['other'][index]=True
    for type in type_list:
        if milk_type_df[type][index]:
            milk_type_df['other'][index]=False
            milk_type_df['types'][index].append(type)
    milk_type_df['type'][index]=milk_type_df['types'][index][0]
            
```

```python
heb_milk_type_df=milk_type_df.drop(type_list, axis=1)
heb_milk_type_df
```

```python
types_df=heb_milk_type_df.groupby(['type']).agg({'id':'count'})
types_df=types_df.rename(columns={"id": "count_id"})
types_df
```

```python
heb_types_df=types_df.reset_index(level=['type'])
heb_types_df
```
