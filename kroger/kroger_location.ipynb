```python
import base64
import requests
import json
import pandas as pd
from config import consumer_key,consumer_secret
```

```python
consumer_key_secret = consumer_key+":"+consumer_secret
consumer_key_secret_enc = base64.b64encode(consumer_key_secret.encode()).decode()
```

```python
headersAuth = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'Authorization': 'Basic '+ str(consumer_key_secret_enc)
}

data = {
    'grant_type': 'client_credentials',
    'scope' : 'product.compact'
}

## Authentication request
response = requests.post('https://api.kroger.com/v1/connect/oauth2/token', headers=headersAuth, data=data, verify=True)
j = response.json()

access_token = (j['access_token'])
```

```python
headersAPI = {
    'accept': 'application/json',
    'Authorization': 'Bearer '+access_token
    }


http = "https://api.kroger.com/v1/locations?filter.chai={{"
chain = '"kroger"'
http2 = "}}&filter.zipCode.near="
zip_near=77002
http3 = "&filter.radiusInMiles="
radius=15
http4 = "&filter.limit="
limit=200


url = http+chain+http2+str(zip_near)+http3+str(radius)+http4+str(limit)   

#response = requests.get('https://api.kroger.com/v1/locations?filter.zipCode.near=77002&filter.radiusInMiles=100&filter.chain="KROGER"&filter.limit=200',headers=headersAPI, verify=True)

response = requests.get(url,headers=headersAPI, verify=True).json()           
#url
```
